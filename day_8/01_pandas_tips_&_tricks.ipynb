# 16 - Spliting a String into Multiple Columns

```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    'name': ['Pooja Dinani', 'Sahiba Memon', 'Nimra Ahmed'],
    'location': ['Gul Center, PK', 'Qasimabad, PK', 'Latifabad, PK']
})
df
```

```python
df.name.str.split(' ').head()
```

```python
df[['first_name', 'last_name']] = df.name.str.split(' ', expand=True)
df
```

```python
df[['city', 'country']] = df['location'].str.split(', ', expand=True)
df
```

```python
df = df[['first_name', 'last_name', 'city', 'country']]
df
```

# 23 - Reshaping a Dataframe

```python
distance = pd.DataFrame([
    ['12345', 100, 200, 300],
    ['34567', 400, 500, 600],
    ['67890', 700, 800, 900]
], columns=['zip', 'factory', 'warehouse', 'retail'])
distance.head()
```

```python
distance.head().T
```

```python
distance2 = pd.DataFrame([
    [1, '12345', 'factory'],
    [2, '34567', 'warehouse'],
], columns=['id', 'zip', 'location_type']
)
distance2.head()
```

```python
distance
```

```python
sns.barplot(x='zip', y='factory', data=distance)
```

```python
distance_melted = distance.melt(id_vars='zip', var_name='location_type', value_name='distance')
distance_melted.head()
```

```python
sns.barplot(x='zip', y='distance', hue='location_type', data=distance_melted)
```
