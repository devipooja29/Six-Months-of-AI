# Plotting with Seaborn

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Important things for plotting as per grammar of graphics

1. Data (df)
2. Types of plot
3. y-axis
4. x-axis
5. Grouping (hue/color) based on categorical variable
6. Size of the plot
7. Color palletes of the plot
8. Style of the plot
9. Interpretation

```python
# Data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Scatter plot
sns.set_theme(style='darkgrid', font='Arial', font_scale=1.5)
plt.figure(figsize=(12,8))
pallete = {'male': 'blue', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=pallete)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter plot of age and fare')
plt.savefig('scatter_plot.png', dpi=300)
plt.savefig('scatter_plot.svg', dpi=300)
```

## *Interpretation*
The scatter plot was drawn to examine the relationship between age and fare of the passengers travelled in Titanic based on their genders. However, no obvious trend has been observed through scatter plot.

```python
# Scatter plot
sns.set_theme(style='darkgrid', font='Arial', font_scale=1.5)
plt.figure(figsize=(12,8))
pallete = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=pallete)
plt.xlabel('sex')
plt.ylabel('fare')
plt.title('Scatter plot of age and fare')
plt.savefig('scatter_plot.png', dpi=300)
plt.savefig('scatter_plot.svg', dpi=300)
plt.legend(loc='upper right', title='Class Rank')
```

```python
# Scatter plot
sns.set_theme(style='darkgrid', font='Arial', font_scale=1.5)
plt.figure(figsize=(16,6))
pallete = {'male': 'blue', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='fare', palette=pallete)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter plot of age and fare')
plt.savefig('scatter_plot.png', dpi=300)
plt.savefig('scatter_plot.svg', dpi=300)
```

```python
# Scatter plot
sns.set_theme(style='darkgrid', font='Arial', font_scale=1.5)
plt.figure(figsize=(12,8))
pallete = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=pallete)
plt.xlabel('sex')
plt.ylabel('fare')
plt.title('Boxen plot of age and fare')
plt.legend(loc='upper right', title='Class Rank')
plt.xticks(rotation=45)
```

```python
# Scatter plot
sns.set_theme(style='darkgrid', font='Arial', font_scale=1.5)
plt.figure(figsize=(12,8))
pallete = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=pallete, capsize=0.1)
plt.xlabel('sex')
plt.ylabel('fare')
plt.title('Barplot of age and fare')
plt.legend(loc='upper right', title='Class Rank')
plt.xticks(rotation=45)
```
