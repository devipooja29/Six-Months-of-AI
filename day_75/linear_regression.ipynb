# Linear Regression

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the data
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
```

```python
# Define X and y (Simple linear regression
# 
# 
# )
X = df[['total_bill']]
y = df['tip']
```

```python
# Define X and y (Multilinear Regression)
X = df[['total_bill', 'size']]
y = df['tip']
```

```python
# Spilt the data into train test spilt
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Call the model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Print the slope, intercept and equation of the model
print('Slope:', model.coef_)
print('Intercept:', model.intercept_)
print('Equation: y = {:.2f}x + {:.2f}'.format(model.coef_[0], model.intercept_))
```

```python
# Predict y
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# Plot the original vs predict values
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
```
