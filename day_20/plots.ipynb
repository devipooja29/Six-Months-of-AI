# Data Visulization

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='#283C43')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.xlabel('Sex')
plt.ylabel('Average Age')
```

```python
sns.barplot(data=df, x='sex', y='fare')
```

```python
sns.barplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.boxplot(data=df, x='sex', y='age', hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age', hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age', hue='sex')
```

```python
sns.scatterplot(data=df, y='fare', x='age', hue='class')
```
