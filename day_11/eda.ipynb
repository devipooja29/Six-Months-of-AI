# *Feature Engineering*

```python
bins = [0,1,5,12,18,30,50,80]
labels = ['Infant', 'Toddler', 'Child', 'Teens', 'Young Adults', 'Adults', 'Old']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```
