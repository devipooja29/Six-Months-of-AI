# Inverse Encoding / Inverse Transform / Decoding Categorical Variables

# Label Encoding

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Impute missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
df.drop('deck', axis=1, inplace=True)
```

```python
# Check missing values 
df.isnull().sum()
```

```python
# Check datatypes
df.dtypes
```

```python
# Encode all categorical and object variables
le_sex = LabelEncoder()
df['sex'] = le_sex.fit_transform(df['sex'])

le_embarked = LabelEncoder()
df['embarked'] = le_embarked.fit_transform(df['embarked'])

le_class = LabelEncoder()
df['class'] = le_class.fit_transform(df['class'])

le_who = LabelEncoder()
df['who'] = le_who.fit_transform(df['who'])

le_embark_town = LabelEncoder()
df['embark_town'] = le_embark_town.fit_transform(df['embark_town'])

le_alive = LabelEncoder()
df['alive'] = le_alive.fit_transform(df['alive'])

df.head()
```

```python
# Inverse transform the data
df['sex'] = le_sex.inverse_transform(df['sex'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df['embark_town'] = le_embark_town.inverse_transform(df['embark_town'])
df['alive'] = le_alive.inverse_transform(df['alive'])
df.head()
```

# One Hot Encoding

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Identify the categorical columns to encode
cat_columns = ['sex', 'embarked']

# Perform One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False)

# Apply encoding to the selected categorical columns
encoded_array = encoder.fit_transform(df[cat_columns])

# Create a DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(cat_columns))

# Concatenate the encoded columns to the original DataFrame and drop the original categorical columns
df_encoded = pd.concat([df, encoded_df], axis=1)
df_encoded.drop(cat_columns, axis=1, inplace=True)

# Display the DataFrame after encoding
df_encoded.head()
```

```python
# Create a copy of the DataFrame for the inverse transformation
df_restored = df_encoded.copy()

# Drop the one-hot encoded columns
encoded_columns = encoder.get_feature_names_out(cat_columns)
df_restored.drop([col for col in encoded_columns if col in df_restored.columns], axis=1, inplace=True)

# Re-add the original categorical columns in their original position
# First, find the position of the 'sex' and 'embarked' columns in the original DataFrame
original_column_order = df.columns.tolist()

# Insert the original categorical columns at their correct positions
for col in cat_columns:
    df_restored.insert(original_column_order.index(col), col, df[col])

# Display the restored DataFrame
df_restored.head()
```
